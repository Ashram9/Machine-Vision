In the next section, you are going to build a linear approximation to the data-manifold in the latent space of the autoencoder. When you walk along this manifold, you will be able to smoothly interpolate between digits, effectively enforcing a smooth ordering on the data.

````python
# Create sampling grid in a 2D subspace
lim = 3
image_dims =int(np.ceil(np.sqrt(n_samples)))
lin_range = np.linspace(-lim,lim,image_dims)
X, Y = np.meshgrid(lin_range, lin_range)
sampling_grid = np.concatenate([np.reshape(Y,[n_samples,1]), np.reshape(X,[n_samples,1])], axis=1)
````

````python
# Encoder encode 784D data to a 500D latent space
# There is a manifold in 500D latent space. What we do is trying to find a approximation of the true manifold (unknown) by PCA.
# We use top-2 principle direction to represent the manifold
# 'sampling_grid' is a set of coordinates we use to interpolate digits on the manifold
````

````python
latent_code, __ = mlp_forward(encoder, mnist_images)
````

````python
def pca(X, num_components=2):
    # Center the data
    mean_X = np.mean(X, axis=0)
    X_centered = X - mean_X
    
    # Compute the covariance matrix
    covariance_matrix = np.cov(X_centered, rowvar=False)
    
    # Compute the eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eigh(covariance_matrix)
    
    # Sort eigenvectors by decreasing eigenvalues
    indices = np.argsort(eigenvalues)[::-1]
    eigenvectors = eigenvectors[:, indices]
    
    # Keep the top num_components eigenvectors
    top_eigenvectors = eigenvectors[:, :num_components]
    
    # Project the data onto the principal components
    principal_components = X_centered @ top_eigenvectors
    
    return principal_components, top_eigenvectors
````

## 4.1) TODO: Compute the covariance matrix of the latent codes

````python
mean = np.mean(latent_code, axis=0)
latent_cov = (latent_code - mean).T @ (latent_code - mean) / (latent_code.shape[0] - 1)
````

## 4.2) TODO: Do PCA on the sampled code
Perform the SVD on the covariance matrix and retain the first two rows of Vt (as a column)

````python
# Perform SVD on the covariance matrix
U, S, Vt = np.linalg.svd(latent_cov)

# Retain the first two rows of V as columns
principal_subspace = Vt[:2, :]

latent_grid = sampling_grid @ principal_subspace
# Reconstruct images from the codes
reconstruction, __ = mlp_forward(decoder, latent_grid)

plt.figure(4, figsize=(6,6))
plot_tiled_array(reconstruction, 'Fitted subspace')
````
